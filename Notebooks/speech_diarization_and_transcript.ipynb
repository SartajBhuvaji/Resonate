```python
import whisper
import dotenv
import pandas as pd
import os
from pyannote.audio import Pipeline
from utils import diarize_text

dotenv.load_dotenv()
HUGGINGFACE_TOKEN = os.getenv("HUGGINGFACE_TOKEN")
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization-3.1",
                                    use_auth_token=HUGGINGFACE_TOKEN)
```

```python
model = whisper.load_model("tiny.en")
file = "mini_ami.wav"

asr_result = model.transcribe(file)
diarization_result = pipeline(file)
final_result = diarize_text(asr_result, diarization_result)

df = pd.DataFrame()
for seg, spk, sent in final_result: 

    df = df.append({'start': seg.start, 'end': seg.end, 'speaker': spk, 'text': sent}, ignore_index=True)
    line = f'{seg.start:.2f} {seg.end:.2f} {spk} {sent}'
    print(line)

df.to_csv(f'speech_transcript_{file}.csv', index=False)    
```
